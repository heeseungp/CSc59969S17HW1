```python
#type of fruits
orangestotal = '   Oranges, Total'
othercitrusfruit = '   Other citrus fruit'
stonefruit = '   Stone fruit'
tropicalfruit = '   Tropical fruit'
applesasfruit = '     Apples as fruit'
bananas = '   Bananas'
berries = '   Berries'
grapes = '   Grapes'
melons = '   Melons'
fruitList = [applesasfruit, bananas, berries, grapes, melons, orangestotal, othercitrusfruit, stonefruit, tropicalfruit]
fruitNameList = []
for i in fruitList:
    fruitNameList.append(str(i))
```

```python
#type of dairy
fluidmilktotal = '   Fluid milk, Total'
butter = '   Butter'
cheese = '   Cheese'
yogurt = '   Yogurt'
dairyother = '   Dairy, Other'
dairyList = [fluidmilktotal, butter, cheese, yogurt, dairyother]
dairyNameList = []
for i in dairyList:
    dairyNameList.append(str(i))
```

```python
#using .loc to get the specific cell of the data and casting to float to convert to float
print(float(mfah94.loc[applesasfruit]))
```

```python
#extracts by any type + gender 
def extractByType(genderExcel, typeList):
    final = []
    temp = []
    for i, typeList_element in enumerate(typeList):
        for j, genderExcel_element in enumerate(genderExcel):
            temp.append(float(genderExcel_element.loc[typeList_element]))
        final.append(temp)
        temp = []
    return(final)
```

```python
menFinalFruit = extractByType(menList, fruitList)
womenFinalFruit = extractByType(womenList, fruitList)
menFinalDairy = extractByType(menList, dairyList)
womenFinalDairy = extractByType(womenList, dairyList)
```

```python
def menFruit(menFinalFruit, fruitNameList, xaxis):
    for idx, element in enumerate(menFinalFruit):
        plt.plot(xaxis, element, str(lineColor[idx%7]) + str(lineStyle[idx%2]), label = fruitNameList[idx])
    plt.axis([0,5,    40,100])

    plt.xlabel('Years')
    plt.xticks(xaxis, yrs)
    plt.ylabel('Percentage')
    plt.legend(loc=3,prop={'size':6})
    plt.title('Men\'s Consumption Percentage of Particular Types of Fruit vs. Time')
    plt.figtext(.1,-0.1, "From this graph, men are consistently consuming less and less citrus fruits each year.")
    plt.figtext(.1,-0.15, "We also observe that tropical fruit's consumption percentage is rapidly growing compared to other fruit's consumption percentage.")
    plt.show()
    
def womenFruit(womenFinalFruit, fruitNameList, xaxis):
    for idx, element in enumerate(womenFinalFruit):
        plt.plot(xaxis, element, str(lineColor[idx%7]) + str(lineStyle[idx%2]), label = fruitNameList[idx])
    plt.axis([0,5,    40,100])

    plt.xlabel('Years')
    plt.xticks(xaxis, yrs)
    plt.ylabel('Percentage')
    plt.legend(loc=3,prop={'size':6})
    plt.title('Women\'s Consumption Percentage of Particular Types of Fruit vs. Time')
    plt.figtext(.1,-0.1, "From this graph, women did consume less citrus fruits in 94-98, but consumped more and more later in 05-08. ")
    plt.figtext(.1,-0.15, "We see big fluctuation in line graph from berries, other citrus fruit, and tropical fruits.")
    plt.show()
    
def menDairy(menFinalDairy, dairyNameList, xaxis):    
    for idx, element in enumerate(menFinalDairy):
        plt.plot(xaxis, element, str(lineColor[idx%7]) + str(lineStyle[idx%2]), label = dairyNameList[idx])
    plt.axis([0,5,    40,100])

    plt.xlabel('Years')
    plt.xticks(xaxis, yrs)
    plt.ylabel('Percentage')
    plt.legend(loc=3,prop={'size':6})
    plt.title('Men\'s Consumption Percentage of Particular Types of Dairy vs. Time')
    plt.figtext(.1,-0.1, "From this graph, women did consume less citrus fruits in 94-98, but consumped more and more later in 05-08. ")
    plt.figtext(.1,-0.15, "From this observation, we also see that there has not been a major change in consumption of cheese.")
    plt.show()
    
def womenDairy(womenFinalDairy, dairyNameList, xaxis):
    for idx, element in enumerate(womenFinalDairy):
        plt.plot(xaxis, element, str(lineColor[idx%7]) + str(lineStyle[idx%2]), label = dairyNameList[idx])
    plt.axis([0,5,    40,100])

    plt.xlabel('Years')
    plt.xticks(xaxis, yrs)
    plt.ylabel('Percentage')
    plt.legend(loc=3,prop={'size':6})
    plt.title('Women\'s Consumption Percentage of Particular Types of Dairy vs. Time')
    plt.figtext(.1,-0.1, "From this graph, we see that yogurt and butter consumption percentage goes up and down together throughout the timeline.")
    plt.figtext(.1,-0.15, "From this observation, we see that fluid milk total did not show a major change in the percentage of consumption.")
    plt.show()
```

```python
lineStyle = ['-', '--', '-.', ':', '.', ',','o','v', '^']
lineColor = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
xaxis = [1,2,3,4]

menFruit(menFinalFruit, fruitNameList, xaxis)
```
